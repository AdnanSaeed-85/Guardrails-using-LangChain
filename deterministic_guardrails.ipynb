```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
from CONFIG import GROQ_MODEL
```

```python
load_dotenv()
llm = ChatGroq(model=GROQ_MODEL)
```

```python
def deterministic_approach(text):
    harms = ['bomb', 'neuclear', 'kill', 'hack']
    return any(t in text.lower() for t in harms)

text_input = [
    'how to make neuclear bomb',
    'what do you think about AI',
    'how to hack server',
    'is AI safe for humanity?',
    'is neuclear safe for humanity?',
    'i will kill you',
    'will ai kill the human jobs?'
]

def printed(text):
    output = deterministic_approach(text)
    return 'Blocked' if output else 'Allowed'

for i in text_input:
    output = printed(i)
    print(f"{output}: {i}")
```
