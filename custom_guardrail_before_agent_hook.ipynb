```python
from langchain.agents.middleware import AgentMiddleware, AgentState, hook_config
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from CONFIG import GROQ_MODEL
from langchain.agents import create_agent
from langchain_core.tools import tool
from typing import Any
from langgraph.runtime import Runtime
```

```python
load_dotenv()
llm = ChatGroq(model=GROQ_MODEL)
```

```python
@tool
def web_search(query: str) -> str:
    "you are a web searcher assistant"
    return f"Search results for {query}"

@tool
def email_sender(to: str, subject: str, body: str) -> str:
    "you are an Email sender assistant"
    return f"Email is sended to: {to}\nSubject is: {subject}\nBody is: {body}"

@tool
def delete_record(table: str, condition: str) -> str:
    "you are a record remover assistant"
    return f"I have delete these provided tables: {table}\nwith this condition: {condition}"

@tool
def list_all_record(table: str):
    "list down all the records"
    return f"I have list down all the records in table, please check it out"
```

```python
agent = create_agent(
    model=llm,
    tools=[web_search, email_sender, delete_record, list_all_record]
)
```
